##5. Set Up Great Expectations Context and Data Source

```python


context = gx.get_context()
data_source = context.data_sources.add_pandas("pandas")
data_asset = data_source.add_dataframe_asset(name="pd dataframe asset")

```

##6. Define and Create a Data Batch

```python
batch_definition = data_asset.add_batch_definition_whole_dataframe(name="my_batch_definition")

batch = batch_definition.get_batch(batch_parameters={"dataframe": df})
```

### Expectation 1

```python
expectation_1 = gx.expectations.ExpectColumnValuesToBeUnique(
    column="Timestamp"
)
```

### Validate Data Against Expectation 1

```python
batch.validate(expectation_1)
```

### Expectation 2

```python
expectation_2 = gx.expectations.ExpectColumnValuesToNotBeNull(column="Timestamp")
```

### Validate Data Against Expectation 2

```python
batch.validate(expectation_2)
```

### Expectation 3

```python
expectation_3 = gx.expectations.ExpectColumnValuesToBeOfType(
    column="Timestamp",
    type_="int64"
)
```

### Validate Data Against Expectation 3

```python
batch.validate(expectation_3)
```
